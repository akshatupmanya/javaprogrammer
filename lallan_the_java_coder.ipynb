```python
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the main content of the page (usually inside <p> tags)
        paragraphs = soup.find_all('p')

        # Extract text from paragraphs and concatenate them
        text_content = '\n'.join([paragraph.get_text() for paragraph in paragraphs])

        return text_content
    else:
        print("Failed to retrieve the Wikipedia page. Status code:", response.status_code)
        return None

# URL of the Wikipedia page you want to scrape
wiki_url = 'https://www.w3schools.com/java/'

# Call the function to scrape the Wikipedia page
content = scrape_wikipedia(wiki_url)
```

```python
import os
import textwrap
import PyPDF2

def extract_text_from_pdf(pdf_path, start_page=13):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(start_page - 1, len(reader.pages)):
            page = reader.pages[page_num]
            page_text = page.extract_text()
            # Remove any extra newlines and spaces
            page_text = ' '.join(page_text.split())
            text += page_text + '\n'
    return text

def align_text(text):
    wrapped_lines = []
    for line in text.split('\n'):
        wrapped_lines.extend(textwrap.wrap(line, width=80))  # Adjust the width as needed
    return '\n'.join(wrapped_lines)

def process_pdfs(pdf_paths):
    all_text = ""
    for pdf_path in pdf_paths:
        if os.path.exists(pdf_path):
            extracted_text = extract_text_from_pdf(pdf_path)
            text = align_text(extracted_text)
            all_text += f"{text}\n\n"
        else:
            all_text += f"PDF file not found: {pdf_path}\n\n"

    return all_text

# Example usage:
pdf_paths = ["/content/javanotes5.pdf"]  # Replace with paths to your PDF files
output_text = process_pdfs(pdf_paths)

# Now you can use the output_text variable as needed.
# For example, you can print it:
print(output_text)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b.T) / (np.linalg.norm(a, axis=1)[:, np.newaxis] * np.linalg.norm(b, axis=1))
```
